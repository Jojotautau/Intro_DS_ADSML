# H. Construction of a boolean matrix.
1. First, I apply a boolean condition on each observation. E.g. for each country, I ask if its region is Africa. I store the object resulting from this boolean condition in a new variable. I also define the object as a matrix instead of an array in order to be able to use the "transpose" operation later. The code looks like this:
```python

```
2. It returns an array of boolean results "true" or "false". I change the format of these boolean results from 'string' to 'integer'.Before I use "transpose()", I have a row-vector whose only line corresponds to the region "Africa" and each columns corresponds to a country. If I transpose this row-vector in a column-vector, the only column correspond to region "Africa" and each line corresponds to a country. This shape match the format required by the exercise. The code looks like this:
```python
a1=a.astype('int').transpose()
```
